Ta cần giải phương trình bậc 4 dạng trùng phương:
ax4 +bx2 +c=0

Trong đó:

Input: 3 hệ số thực a, b, c nhập từ bàn phím.

Output: số nghiệm thực và giá trị nghiệm (nếu có).

Các trường hợp đặc biệt

a = b = c = 0 → phương trình vô số nghiệm.

a = b = 0, c ≠ 0 → phương trình vô nghiệm.

a = 0, b ≠ 0 → phương trình trở thành bậc 2 theo x² → có thể có 0, 2 nghiệm.

a ≠ 0 → giải phương trình bậc 2 theo y = x².

Nếu delta < 0 → vô nghiệm.

Nếu delta ≥ 0 → xét nghiệm y1, y2.

Mỗi nghiệm y >= 0 sinh ra 2 nghiệm x = ±√y.
