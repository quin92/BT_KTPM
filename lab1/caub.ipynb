### Phân tích số lượng test-cases cần thiết
Hàm có 2 điều kiện rẽ nhánh:

Điều kiện 1: x > 10

Điều kiện 2: x > 0 (chỉ xét khi x <= 10)

⇒ Tạo ra 3 nhánh cần kiểm thử:

Nhánh 1: x > 10 → return 2*x

Nhánh 2: 0 < x <= 10 → return -x

Nhánh 3: x <= 0 → return 2*x (sai logic, phải là -x)

Bộ test-case đã xây dựng từ câu a bao gồm cả giá trị biên (x = 0, x = 10) và giá trị âm (x = -5), nên có thể phát hiện lỗi.

#### Số test-case tối thiểu

Cần ít nhất 3 test-case để bao phủ đủ các nhánh.

Tương tự câu a, để phát hiện lỗi tiềm ẩn ở ranh giới điều kiện, ta cần thêm test-case tại:

x = 10 (ranh giới giữa nhánh 1 và nhánh 2)

x = 0 (ranh giới giữa nhánh 2 và nhánh 3)

### Các test-case cụ thể
| Test Case | Input (x) | Expected Output (câu a) | Actual Output (câu b) | Mục đích kiểm thử                          |
| --------- | --------- | ----------------------- | --------------------- | ------------------------------------------ |
| TC1       | 12        | 24                      | 24                    | Nhánh 1: kiểm tra `x > 10`                 |
| TC2       | 8         | -8                      | -8                    | Nhánh 2: kiểm tra `0 < x <= 10`            |
| TC3       | 10        | -10                     | -10                   | Biên giữa nhánh 1 và nhánh 2               |
| TC4       | 0         | 0                       | 0                     | Biên giữa nhánh 2 và nhánh 3               |
| TC5       | -5        | 5                       | -10                   | Nhánh 3: `x < 0` → **phát hiện lỗi logic** |



##### Kết luận:

Cần 5 test-case để kiểm thử đầy đủ các nhánh và các giá trị biên.

Bộ test-case đã xây dựng đủ mạnh để phát hiện lỗi trong phiên bản sai của hàm f1.

Cụ thể, test-case x = -5 cho thấy Expected = 5, nhưng hàm b trả về Actual = -10.

Do đó, bộ test-case đã phát hiện được lỗi logic của hàm.
