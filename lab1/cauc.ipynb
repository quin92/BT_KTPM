#### Phân tích số lượng test-cases cần thiết

Có 2 điều kiện rẽ nhánh:

if (x < 10)

else if (x < 2) (điều kiện này không bao giờ được thực thi, vì khi đã vào else if thì chắc chắn x >= 10, trong khi điều kiện lại yêu cầu x < 2 → mâu thuẫn).

Do đó:

Nhánh 1: x < 10 → return 2*x

Nhánh 2: x >= 10 → return 2*x

Nhánh 3 (else if (x < 2)) là dead code, không thể nào xảy ra.

Vì thực tế hàm chỉ còn 2 nhánh khả thi, số test-case tối thiểu cần thiết = 2.

Test-case 1: chọn giá trị x < 10, ví dụ x = 8 → kiểm thử nhánh 1.

Test-case 2: chọn giá trị x >= 10, ví dụ x = 12 → kiểm thử nhánh 2.

Tuy nhiên, để kiểm tra kỹ hơn ở biên, nên thêm:

x = 10 (ranh giới giữa <10 và >=10)

x = 0 hoặc x = -5 (trường hợp âm hoặc bằng 0, vẫn thuộc nhánh <10)

#### Các test-case cụ thể
| Test Case | Input (x) | Output của f2 | Mục đích kiểm thử                      |
| --------- | --------: | ------------: | -------------------------------------- |
| TC1       |        12 |            24 | Nhánh `x >= 10` (else)                 |
| TC2       |         8 |            16 | Nhánh `x < 10`                         |
| TC3       |        10 |            20 | Biên: `x = 10` (thuộc nhánh `x >= 10`) |
| TC4       |         0 |             0 | Trường hợp trong `x < 10` (0)          |
| TC5       |        -5 |           -10 | Trường hợp âm trong `x < 10`           |


Kết luận:

Tối thiểu cần 2 test-case:

Một test với x < 10 (ví dụ x=8)

Một test với x >= 10 (ví dụ x=12)

Nếu muốn bao phủ biên và phát hiện dead code, có thể dùng thêm x = 10, x = 0, x = -5.
