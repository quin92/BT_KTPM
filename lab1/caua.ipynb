### Phân tích số lượng test-cases cần thiết
Hàm có 1 điều kiện rẽ nhánh if (x > 10).
→ có 2 nhánh cần kiểm thử: khi x > 10, khi x <= 10
<br> Số test-case tối thiểu = 2 (đủ để bao phủ nhánh)


### Các test-case cụ thể

| Test Case | Input (x) | Expected Output | Mục đích kiểm thử               |
| --------- | --------: | --------------: | ------------------------------- |
| TC1       |        12 |              24 | Nhánh **true** (`x > 10`)       |
| TC2       |         5 |              -5 | Nhánh **false** (`x <= 10`)     |
| TC3       |        10 |             -10 | **Biên**: kiểm `x = 10`         |
| TC4       |         0 |               0 | Giá trị đặc biệt `0`            |
| TC5       |        -5 |               5 | Số âm (`x < 0`) kiểm độ bao phủ |

Mặc dù chỉ cần 2 test-case để bao phủ nhánh, nhưng nên bổ sung sử dụng 3 test-case (bao gồm giá trị biên x = 10) để tăng độ tin cậy.
